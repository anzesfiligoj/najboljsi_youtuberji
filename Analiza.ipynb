V tem projektu bom analiziral najpopularnejše youtuberje glede na spletno stran Socialblade.

Podatki, ki sem jih zbral so: Socialblade rang, Ocena, Ime, Število objavljenih videoposnetkov, Število naročnikov
in Število ogledov.
Analiziral bom odnose med: Socialblade rangom ter Številom naročnikov, Socialblade rangom ter Številom ogledov,
Oceno in Številom naročnikov..

```python
#za večjo preglednost pokažemo največ 20 vrstic naenkrat
pd.options.display.max_rows = 20
```

```python
import pandas as pd
%matplotlib inline

#preberemo datoteko
yt = pd.read_csv("csv_datoteka.txt", index_col = "Socialblade_rang")
```

```python
yt
```

```python
yt["Število_ogledov"].sort_values(ascending=False)
```

Ker smo youtuberje zgoraj uredili padajoče glede na število ogledov, je očitno, da ima najbolje rangiran tudi največ ogledov. Če primerjamo največje in najmanjše število ogledov, opazimo, da sta številki skoraj neprimerljivi. zato bom tabeli dodal še en stolpec kjer bom število ogledov youtuberja izrazil v procentih ogledov najboljšega (T-Series) in to število zaradi preglednosti še zaokrožil.

```python
yt["Procent_ogledov_najboljšega"] = round(yt["Število_ogledov"] * 100 / 20265848867)
yt
```

```python
yt.groupby("Procent_ogledov_najboljšega").size().plot(figsize=(20, 10))
```

Očitno se večina youtuberjev zadržuje nekje v spodnjih 10%. To pomeni, da so vodilni v številu ogledov (glej tabelo nekaj celic više) nekakšen fenomen, še posebej T-Series. Večina izmed 5000 youtuberjev s spletne strani Socialblade jim "ne seže niti do kolen". 

```python
yt["Število_naročnikov"].sort_values(ascending=False)
```

Iz zgornje tabele je razvidno, da se številke razlikujejo še v večji meri kot pri številu ogledov, zato se bom težave lotil podobno kot zgoraj.

```python
yt["Procent_naročnikov_najboljšega"] = round(yt["Število_naročnikov"] * 100 / 3562485550)
yt["Procent_naročnikov_najboljšega"].sort_values(ascending=False)
```

```python
yt.groupby("Procent_naročnikov_najboljšega").size().plot(figsize=(20, 10))
```

Ta graf je še bolj nepregleden od prvega, toda pove nam veliko. Že iz tabele v zgornji celici vidimo, da imajo le prvi trije youtuberji vrednost Pregledno_število_naročnikov večjo od 50. Povprečen youtuber pa ima to vrednost zelo blizu 0 oz. zaradi zaokroževaja morda celo 0.
